### *Simple Linear Regression Assignment*

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
import warnings
warnings.filterwarnings("ignore")
```

```python
df = pd.read_csv("My 2026 strava run.csv")
df.head()
```

```python
df.columns
```

```python
df.shape
```

```python
df_lr = df.loc[:, ["Moving Time in Minutes",'Distance']]
```

```python
df_lr.head()
```

```python
# Checking for whether these two variable have any linear relationship
plt.scatter(df_lr["Moving Time in Minutes"], df_lr["Distance"])
plt.show()
```

```python
# Training Model
simple_lr = LinearRegression()
simple_lr.fit(df_lr[["Moving Time in Minutes"]], df_lr["Distance"])
```

```python
print("Slope:", simple_lr.coef_)
```

```python
print("Intercept:", simple_lr.intercept_)
```

```python
plt.style.use('seaborn-v0_8-whitegrid')
plt.scatter(df_lr["Moving Time in Minutes"], df_lr["Distance"], color= "skyblue", alpha= 0.7, edgecolors= "black")
plt.plot(df_lr["Moving Time in Minutes"], simple_lr.predict(df_lr[["Moving Time in Minutes"]]), color='red')
plt.title("Linear Regression", fontsize=16, weight='bold')
plt.xlabel("Moving Time (min)", fontsize= 12)
plt.ylabel("Distance (km)", fontsize= 12)
plt.grid(True, linestyle='--', alpha=0.4)
plt.show()
plt.savefig("Linear Regression Plot")
```

```python
# Prediction of Model if I run for 200 minutes
print("Model prediction if I run for 200 minutes:\n", simple_lr.predict([[250]]))
```
